## Bring in Tree Data

```python
# read in tree shapefile
trees_shape = gpd.read_file("data/2015StreetTreeCensus/geo_export_fb0b795e-20cf-488d-9d82-905967f34e5a.shp")
```

```python
trees_shape.head()
```

```python
trees_shape.columns
```

```python
# check NTA
print(trees_shape['nta'].drop_duplicates().to_list())
```

```python
# get count of unique NTAs
trees_shape['nta'].value_counts()
```

```python
# check CRS
trees_shape.crs
```

```python
# re-set CRS
trees_shape = trees_shape.to_crs(6933)
```

```python
# recheck CRS
trees_shape.crs
```

### Spatial Join: Left + Within (keep Trees on left)

```python
# left join - keep trees on the left... to retain tree geometry and keep all trees

trees_nta = gpd.sjoin(trees_shape, nyc_ntas[['ntacode', 'ntaname', 'geometry']], how = 'left', op = 'within')
trees_nta.head()
```

```python
trees_nta.shape
```

### Count Number of Trees per NTA

```python
# create a new DF - count the number of trees by counting on GEOID
trees_per_nta = pd.DataFrame(trees_nta['ntacode'].value_counts())
```

```python
# rename column
trees_per_nta.columns = ['tree_count']
```

```python
# set GEOID as a proper column, not index, so we can merge in it
trees_per_nta['ntacode'] = list(trees_per_nta.index)
```

```python
trees_per_nta = trees_per_nta.reset_index(drop = True)
```

```python
trees_per_nta.head()
```

### Merge NTAs (with Polygons) with Tree Counts (this is only info we need from trees)

```python
# keep census data on left as to retain polygon geometry
nta_trees = pd.merge(nyc_ntas, trees_per_nta, how = 'left', on = 'ntacode')
```

```python
# get area per square mile for each census tract, in order to calculate on
# have to convert to miles, as CRS is in meters
nta_trees['area_per_sqmile'] = nta_trees['geometry'].area / 2.59e+6
```

```python
# create column to get trees per square mile
nta_trees['trees_per_sqmile_nta'] = nta_trees['tree_count'] / nta_trees['area_per_sqmile']
```

```python
nta_trees.shape
```

```python
nta_trees.head()
```

```python
# initial plot to ensure we have all data - plot tree per square mile

# create a figure with one subplot
fig, ax = plt.subplots(figsize=(20,10))

# plot census tracts
nta_trees.plot(ax = ax, alpha = 0.3, edgecolor = 'black', linewidth = 0.5, color = 'gray')

# plot tree pop
ax = nta_trees.plot(ax = ax, column = 'trees_per_sqmile_nta', cmap= "Greens", alpha = 0.7, linewidth = 0.5, legend = True)

ax.axis('off')
ax.set(title = 'Number of Trees per Square Mile - by NTA (2010)')

plt.savefig("maps/5Borough/NTA-Trees-per-Square-Mile.pdf")
```

```python
# export file

nta_trees.to_file("maps/MapData/nyc-ntas.shp")
```
